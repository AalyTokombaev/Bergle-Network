```python
import pandas as pd
import h3
```

```python
csv_file = "Boligstroek.csv"

df = pd.read_csv(csv_file)
```

```python
df
```

# Calculate maximum distance

```python
def calculate_distance(coords_1, coords_2):
    return h3.point_dist(coords_1, coords_2, unit='m')
```

```python
distances = {}

# Iterate through each row to calculate distances
for index_1, row_1 in df.iterrows():
    for index_2, row_2 in df.iterrows():
        # Skip calculating distance for the same pair of coordinates
        if index_1 != index_2:
            coords_1 = (row_1['long'], row_1['lat'])
            coords_2 = (row_2['long'], row_2['lat'])
            distance = calculate_distance(coords_1, coords_2)

            # Store the distance in the dictionary
            distances[(row_1["stroek_navn"], row_2["stroek_navn"])] = distance
```

```python
distance_df = pd.DataFrame.from_dict(distances, orient='index', columns=['distance'])
```

```python
distance_df.sort_values(by="distance",ascending=False)
```
